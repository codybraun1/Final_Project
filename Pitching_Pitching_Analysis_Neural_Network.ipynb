Pitching Stats - Neural Network

```python
from pybaseball import pitching_stats_bref
import pandas as pd
pitch_bref = pitching_stats_bref(2008)
mvp_playerid_full = pd.read_csv('MVP_Players_Info.csv')
mvp_playerid_full['Name']=mvp_playerid_full['nameFirst'] + " " + mvp_playerid_full['nameLast']
mvp_playerid_full.head()
mvp_player_pitch_bref = pitch_bref.merge(mvp_playerid_full[['Name', 'awardID']], right_on = 'Name', left_on = 'Name', how='left', suffixes=('_mvp', '_pitch'))
```

```python
import numpy as np
mvp_player_pitch_bref['MVP_Classification']=mvp_player_pitch_bref['awardID']
mvp_player_pitch_bref['MVP_Classification'] = mvp_player_pitch_bref['MVP_Classification'].map({'MVP': 1})
mvp_player_pitch_bref['MVP_Classification'] = mvp_player_pitch_bref['MVP_Classification'].replace(np.nan, 0)
```

```python
 from sklearn.model_selection import train_test_split
y=mvp_player_pitch_bref['MVP_Classification']
X=mvp_player_pitch_bref.drop(['MVP_Classification','awardID', 'Name','Lev', 'Tm'], axis=1)
X = X.fillna(value=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from keras.utils import to_categorical

# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
y_train_categorical
```

```python
from keras.models import Sequential

model = Sequential()
```

```python
 from keras.layers import Dense
number_inputs = 3
number_hidden_nodes = 4
model.add(Dense(units=number_hidden_nodes,
                activation='relu', input_shape=(37,)))
```

```python
number_classes = 2
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=1000,
    shuffle=True,
    verbose=2
)
```

```python
 # Evaluate the model using the testing data
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f"Pitching Data Loss: {model_loss}, Pitching Model Accuracy: {model_accuracy}")
```
