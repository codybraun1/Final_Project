```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
mvps = pd.read_csv('MVP_Players_Info_Fixed.csv')
mvps["fullName"] = mvps["nameFirst"]+" "+mvps["nameLast"]
mvps
```

```python
from pybaseball import playerid_reverse_lookup
id_lookup = playerid_reverse_lookup(list(mvps['bbrefID']), key_type='bbref')
id_lookup_df = pd.DataFrame(id_lookup[['key_bbref','key_fangraphs']])
id_lookup_df.head()
```

```python
mvps_full = mvps.merge(id_lookup_df, left_on="bbrefID", right_on="key_bbref", how="left")
mvps_full
```

```python
batting = pd.read_csv("BattingDataFG.csv")
pitching = pd.read_csv("PitchingDataFG.csv")
fielding = pd.read_csv("FieldingDataFG.csv")
records = pd.read_csv("TeamRecords.csv")
fielding.head()
```

```python
#merge batting, pitching, fielding
full_1 = batting.merge(pitching, on=["playerid","Season"], how="outer", suffixes=("_bat","_pit"))
full_2 = full_1.merge(fielding[["Season","Pos","FP","playerid"]], on=["playerid","Season"], how="outer", suffixes=("","_fld"))
full_2.head()
```

```python
records.columns
```

```python
full_3 = full_2.merge(mvps_full,left_on=["Season","playerid"],right_on=["yearID","key_fangraphs"],how="outer")
full_3.columns
```

```python
#merge team win% & playoffs
records.loc[(records["DivWin"]=="Y") | (records["WCWin"]=="Y") | (records["LgWin"]=="Y"), "Playoffs"] = 1
records["Playoffs"] = records["Playoffs"].replace(np.nan, 0)

full_df = full_3.merge(records,left_on=["Season","teamID"],right_on=["yearID","teamID"],how="left")
full_df.describe()
```

```python
full_df['MVP_Classification']=full_df['awardID']
full_df['MVP_Classification'] = full_df['MVP_Classification'].map({'MVP': 1})
full_df['MVP_Classification'] = full_df['MVP_Classification'].replace(np.nan, 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
y=full_df.loc[full_df["Season"]==2018]['MVP_Classification']
X=full_df.loc[full_df["Season"]==2018].drop(['MVP_Classification','awardID', 'Name','Lev', 'Tm', 'Age'], axis=1)
X = X.fillna(value=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```
