```python
import pandas as pd
import numpy as np
```

```python
mvps = pd.read_csv('MVP_Players_Info_Fixed.csv')
#mvps["fullName"] = mvps["nameFirst"]+" "+mvps["nameLast"]
mvps
```

```python
from pybaseball import playerid_reverse_lookup
id_lookup = playerid_reverse_lookup(list(mvps['bbrefID']), key_type='bbref')
id_lookup_df = pd.DataFrame(id_lookup[['key_bbref','key_fangraphs']])
id_lookup_df.head()
```

```python
mvps_full = mvps.merge(id_lookup_df, left_on="bbrefID", right_on="key_bbref", how="left")
mvps_full
```

```python
batting = pd.read_csv("BattingDataFG.csv")
pitching = pd.read_csv("PitchingDataFG.csv")
fielding = pd.read_csv("FieldingDataFG.csv")
fielding
```

```python
idx = fielding.groupby(["Season","playerid"])["G"].transform(max) == fielding["G"]
fielding_1 = fielding[idx]
idx = fielding_1.groupby(["Season","playerid","G"])["Inn"].transform(max) == fielding_1["Inn"]
fielding_fixed = fielding_1[idx]
fielding_fixed
```

```python
#merge batting, pitching, fielding
full_1 = batting.merge(pitching, on=["playerid","Season"], how="outer", suffixes=("_bat","_pit"))
full_1
```

```python
full_2 = full_1.merge(fielding_fixed[["Season","Team","Pos","FP","playerid"]], on=["playerid","Season"], how="left", suffixes=("","_fld"))
full_2
```

```python
full_3 = full_2.merge(mvps_full,left_on=["Season","playerid"],right_on=["yearID","key_fangraphs"],how="outer")
full_3
```

```python
team_lookup = pd.read_csv("player_team_lookup.csv")
team_lookup.head()
```

```python
id_lookup_2 = playerid_reverse_lookup(list(team_lookup['playerID']), key_type='bbref')
id_lookup_2_df = pd.DataFrame(id_lookup_2)
id_lookup_2_df.head()
```

```python
team_lookup_2 = team_lookup.merge(id_lookup_2_df[["key_fangraphs","key_bbref"]],left_on="playerID",right_on="key_bbref",how="left")
team_lookup_2.head()
```

```python
full_4 = full_3.merge(team_lookup_2[["yearID","teamID","lgID","key_fangraphs","key_bbref"]],left_on=["Season","playerid"],right_on=["yearID","key_fangraphs"],how="left")
full_4
```

```python
#merge team win% & playoffs
records = pd.read_csv("TeamRecords.csv")
records.loc[(records["DivWin"]=="Y") | (records["WCWin"]=="Y") | (records["LgWin"]=="Y"), "Playoffs"] = 1
records["Playoffs"] = records["Playoffs"].replace(np.nan, 0)

full_df = full_4.merge(records[["yearID","teamID","G","W","L","Playoffs"]],left_on=["Season","teamID_y"],right_on=["yearID","teamID"],how="left",suffixes=["_pit","_team"])
list(full_df.columns)
full_df
```
